```python
data_region2 = pd.DataFrame.from_records(data_region,columns=header)
data_region2
```

```python
data =data_covidupdate.merge(data_region2,how='left',left_on='province_of_onset',right_on='จังหวัด')
data
```

```python
print(data.isnull().any()) ## ตรวจสอบค่า Missing หลังจากรวมตาราง
print('*'* 100)
data2 = data.dropna() #แก้Missingด้วยการ Dropna
print(data2.isnull().any()) ## ตรวจสอบค่า Missing อีกครั้ง
data2
```

```python
data3 = data2.groupby('จังหวัด').size().reset_index(name='ยอดจำนวนผู้ติดเชื้อ') 
data3
```

```python
data4 =data3.merge(data_region2,how='right',left_on='จังหวัด',right_on='จังหวัด')
data4
```
