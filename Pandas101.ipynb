````
female = this_data[this_data['sex']=='หญิง'] #สร้างตัวแปร เลือกเฉพาะที่เป็นเพศหญฺง
female['age'] #พร้อมกับแสดงคอลัมอายุ
````

````
sum = 0 #ให้ sum เป็นตัวแปรโดยกำหนดให้เท่ากับ 0 เพราะถ้าเป็นค่าอื่นเลขมันจะเพิ่มขึ้นตามเลขที่เรากำหนด
N = 0 #เป็นตัวแปรเท่ากับ 0
for a in female['age']:
    
    if a > 0: #ถ้าอายุ(a)มากกว่า 0
        sum += a # sum = sum + a (ผลรวมอายุทั้งหมดมาบวกกับ a เรื่อยๆ) 
        N += 1 #คำนวณจำนวนโดยบวก 1 ไปเรื่อยๆ มี 1 ตัว 2,3,4,5 ตัวบวก 1 ไปเรื่อยๆ

print(f'อายุเฉลี่ย ของ ผู้ป่วยหญิง {sum/N}')
````

````
female_KK = female[female['province_of_onset']=='ขอนแก่น']

sum = 0
N = 0
for a in female_KK['age']:
    
    if a > 0:
        sum += a # sum = sum + a
        N += 1

print(f'อายุเฉลี่ย ของ ผู้ป่วยหญิง ในจังหวัดขอนแก่น {sum/N}')
````

````
this_data = data_covid[['sex','age','province_of_onset']] #เลือกคอลัมที่ต้องการคือ เพศ อายุ และจังหวัด
this_data #แสดงผล
````

````
sum = 0 #มีตัวแปร sum มารับให้เริ่มเท่ากับ 0 ถ้าเริ่มค่าอื่นไม่ได้ /ถ้าเป็นคูณเริ่มที่ 1
N = 0 #เอาตัวเลขทั้งหมด 
for a in male['age']: 
    
    if a > 0: #เลือกมากกว่า 0 เพื่อหลีกเลี่ยงค่าผิดปกติ
        sum += a # sum = sum + a  #เอา a มาบวก sum เรื่อยๆ
        N += 1 

print(f'อายุเฉลี่ย ของ ผู้ป่วยชาย {sum/N}') #อายุเฉลี่ย หาร จำนวน
````

````
data_HW2 = list() #สร้าง list เพื่อวนลูปหา index วันที่ 1-20
for each_row in data_covid.iterrows():  
    if each_row[1]['announce_date'].split('/')[1] <= '20':  #เลือกคอลัม announce_date โดยใช้คำสั่ง split ในการแยกเอาเฉพาะวันที่ โดยค่าต้องมากกว่าหรือเท่ากับ 20
        data_HW2.append(True) #ถ้าวนลูปแล้วเป็น True จะเพิ่มเข้า list
    else:
        data_HW2.append(False) #ถ้าวนลูปแล้วเป็น False จะไม่เพิ่มเข้า list
data_covid[data_HW2]
````
